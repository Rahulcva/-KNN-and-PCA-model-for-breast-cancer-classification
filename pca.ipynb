```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train) # Calculates mean and std dev from X_train and # Transform the training data (mean center and scale)
```

```python
x_test=scaler.transform(x_test)
# Transform the test data using the same scaling parameters from X_train
```

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
pca=PCA(n_components=None) # here we say how many principal component u want  if u give None so all column it take if u set specifially so u can do that
```

```python
pca.fit(x_train)
```

```python
#for finding eigen vector
pca.explained_variance_
```

```python
#for finding eigen value
pca.components_
```

```python
#start we n_component so we know how many n component is important so we for this we apply cummulative sum
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca1=PCA(n_components=10) # here we say how many principal component u want  if u give None so all column it take if u set specifially so u can do that
```

```python
pca1.fit(x_train)
```

```python
pca1.explained_variance_
```

```python
pca1.components_
```

```python
#take 90% of the variance min n_components
```

```python
x_train1=pca1.fit_transform(x_train)
```

```python
x_test1=pca1.transform(x_test)
```
